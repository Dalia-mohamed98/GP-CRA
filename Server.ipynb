### Setup

```
# add path
import sys
sys.path.append("espnet")

# define device
import torch
device = torch.device("cuda")

# define E2E-TTS model
from argparse import Namespace
from espnet.asr.asr_utils import get_model_conf
from espnet.asr.asr_utils import torch_load
from espnet.utils.dynamic_import import dynamic_import
idim, odim, train_args = get_model_conf(model_path)
model_class = dynamic_import(train_args.model_module)
model = model_class(idim, odim, train_args)
torch_load(model_path, model)
model = model.eval().to(device)
inference_args = Namespace(**{
    "threshold": 0.5,"minlenratio": 0.0, "maxlenratio": 10.0,
    # Only for Tacotron 2
    "use_attention_constraint": True, "backward_window": 1,"forward_window":3,
    # Only for fastspeech (lower than 1.0 is faster speech, higher than 1.0 is slower speech)
    "fastspeech_alpha": 1.0,
    })

# define neural vocoder
import yaml
import parallel_wavegan.models
with open(vocoder_conf) as f:
    config = yaml.load(f, Loader=yaml.Loader)
vocoder_class = config.get("generator_type", "ParallelWaveGANGenerator")
vocoder = getattr(parallel_wavegan.models, vocoder_class)(**config["generator_params"])
vocoder.load_state_dict(torch.load(vocoder_path, map_location="cpu")["model"]["generator"])
vocoder.remove_weight_norm()
vocoder = vocoder.eval().to(device)
if config["generator_params"]["out_channels"] > 1:
    from parallel_wavegan.layers import PQMF
    pqmf = PQMF(config["generator_params"]["out_channels"]).to(device)

# define text frontend
from tacotron_cleaner.cleaners import custom_english_cleaners
from g2p_en import G2p
with open(dict_path) as f:
    lines = f.readlines()
lines = [line.replace("\n", "").split(" ") for line in lines]
char_to_id = {c: int(i) for c, i in lines}
g2p = G2p()
def frontend(text):
    """Clean text and then convert to id sequence."""
    text = custom_english_cleaners(text)
    
    if trans_type == "phn":
        text = filter(lambda s: s != " ", g2p(text))
        text = " ".join(text)
        print(f"Cleaned text: {text}")
        charseq = text.split(" ")
    else:
        print(f"Cleaned text: {text}")
        charseq = list(text)
    idseq = []
    for c in charseq:
        if c.isspace():
            idseq += [char_to_id["<space>"]]
        elif c not in char_to_id.keys():
            idseq += [char_to_id["<unk>"]]
        else:
            idseq += [char_to_id[c]]
    idseq += [idim - 1]  # <eos>
    return torch.LongTensor(idseq).view(-1).to(device)

import nltk
nltk.download('punkt')
print("Now ready to synthesize!")
```

### Synthesis

```
import time
from scipy.io.wavfile import write
def MY_text_to_voice(sentence):
  with open(vocoder_conf) as f:
    config = yaml.load(f, Loader=yaml.Loader)
  vocoder_class = config.get("generator_type", "ParallelWaveGANGenerator")
  vocoder = getattr(parallel_wavegan.models, vocoder_class)(**config["generator_params"])
  vocoder.load_state_dict(torch.load(vocoder_path, map_location="cpu")["model"]["generator"])
  vocoder.remove_weight_norm()
  vocoder = vocoder.eval().to(device)
  pad_fn = torch.nn.ReplicationPad1d(config["generator_params"].get("aux_context_window", 0))
  use_noise_input = vocoder_class == "ParallelWaveGANGenerator"
  with torch.no_grad():
    start = time.time()
    x = frontend(sentence)
    c, _, _ = model.inference(x, inference_args)
    c = pad_fn(c.unsqueeze(0).transpose(2, 1)).to(device)
    xx = (c,)
    if use_noise_input:
        z_size = (1, 1, (c.size(2) - sum(pad_fn.padding)) * config["hop_size"])
        z = torch.randn(z_size).to(device)
        xx = (z,) + xx
    if config["generator_params"]["out_channels"] == 1:
        y = vocoder(*xx).view(-1)
    else:
        y = pqmf.synthesis(vocoder(*xx)).view(-1)    
  rtf = (time.time() - start) / (len(y) / config["sampling_rate"])
  print(f"RTF = {rtf:5f}")
  return [y.view(-1).cpu().numpy(), config["sampling_rate"]]

def LJ_text_to_voice(sentence):
  with open(vocoder_conf_lj) as f:
    config = yaml.load(f, Loader=yaml.Loader)
  vocoder_class = config.get("generator_type", "ParallelWaveGANGenerator")
  vocoder = getattr(parallel_wavegan.models, vocoder_class)(**config["generator_params"])
  vocoder.load_state_dict(torch.load(vocoder_path_lj, map_location="cpu")["model"]["generator"])
  vocoder.remove_weight_norm()
  vocoder = vocoder.eval().to(device)
  pad_fn = torch.nn.ReplicationPad1d(config["generator_params"].get("aux_context_window", 0))
  use_noise_input = vocoder_class == "ParallelWaveGANGenerator"
  with torch.no_grad():
    start = time.time()
    x = frontend(sentence)
    c, _, _ = model.inference(x, inference_args)
    c = pad_fn(c.unsqueeze(0).transpose(2, 1)).to(device)
    xx = (c,)
    if use_noise_input:
        z_size = (1, 1, (c.size(2) - sum(pad_fn.padding)) * config["hop_size"])
        z = torch.randn(z_size).to(device)
        xx = (z,) + xx
    if config["generator_params"]["out_channels"] == 1:
        y = vocoder(*xx).view(-1)
    else:
        y = pqmf.synthesis(vocoder(*xx)).view(-1)    
  rtf = (time.time() - start) / (len(y) / config["sampling_rate"])
  print(f"RTF = {rtf:5f}")
  return [y.view(-1).cpu().numpy(), config["sampling_rate"]]
```

```
from tika import parser
import re,string
def split_text(text):
  max_char_per_sentence = 120

  text = text.replace('\n', ' ')
  text = text.replace('•', '.')
  text = text.replace('‖', ' ')
  text = text.replace('                       ', ' ')
  text = text.replace('                      ', ' ')
  text = text.replace('                     ', ' ')
  text = text.replace('                    ', ' ')
  text = text.replace('                   ', ' ')
  text = text.replace('                  ', ' ')
  text = text.replace('                 ', ' ')
  text = text.replace('                ', ' ')
  text = text.replace('               ', ' ')
  text = text.replace('              ', ' ')
  text = text.replace('             ', ' ')
  text = text.replace('            ', ' ')
  text = text.replace('           ', ' ')
  text = text.replace('          ', ' ')
  text = text.replace('         ', ' ')
  text = text.replace('        ', ' ')
  text = text.replace('       ', ' ')
  text = text.replace('      ', ' ')
  text = text.replace('     ', ' ')
  text = text.replace('    ', ' ')
  text = text.replace('   ', ' ')
  text = text.replace('  ', ' ')
  text = text.replace('~', ' nearly ')
  text = text.replace('@', ' at ')
  text = text.replace('#', ' hashtag ')
  text = text.replace('$', ' dollar ')
  text = text.replace('%', 'percent ')
  text = text.replace('&', ' and ')
  text = text.replace('_', ' under score ')
  text = text.replace('*', ' multiply ')
  text = text.replace('-', ' dash ')
  text = text.replace('+', ' plus ')
  text = text.replace('=', ' equal ')
  text = text.replace('/', ' divide ')
  text = text.replace('\\', ' slash ')
  text = text.replace('&', ' and ')
  text = text.replace('||', ' or ')
  text = text.replace('(', ' ')
  text = text.replace(')', ' ')
  text = text.replace('{', ' ')
  text = text.replace('}', ' ')
  text = text.replace('>', ' greater than ')
  text = text.replace('<', ' smaller than ')
  text = text.replace('>=', ' greater than or equal ')
  text = text.replace('<=', ' smaller than or equal ')
  start_index = 0
  current_index = 0
  sentences = []
  for char in text:
    if char in [',','.',';','.','?','؟','!',':']:
      sentences.append(text[start_index:current_index + 1])
      start_index = current_index + 1
    current_index = current_index + 1
  sentences.append(text[start_index:])

  current_index = 0
  for s in sentences:
    if len(s) > max_char_per_sentence:
      #search for space at the middle to split on it
      checking_for_space_index = int(max_char_per_sentence/2)
      while s[checking_for_space_index] != " ":
        checking_for_space_index = checking_for_space_index + 1

      sentences[current_index] = s[0:checking_for_space_index+1]
      sentences.insert(current_index+1,s[checking_for_space_index+1:])
    current_index = current_index + 1

  return sentences
  
def extract_from_pdf(file_path):
  data = ""
  if file_path[-3:] == "pdf":
    raw = parser.from_file(file_path)
    data = raw['content']
  else:
    with open(file_path, 'r') as file:
      data = file.read().replace('\n', '')
  
  print(data) 
  res = split_text(data)
  for sntnc in res:
    if sntnc == " " or sntnc in string.punctuation:
      res.remove(sntnc)
  return res
```

```
from scipy.io.wavfile import write
from pydub import AudioSegment
import os
def write_to_text_file_using_sentences(file_path,sentences):
  f = open(file_path, "w")
  for sentence in sentences:
    f.write(sentence + "\n")
  f.close()

def write_to_text_file_using_audio(file_path,array,s_rate):
  write(file_path,s_rate,array)
  AudioSegment.from_wav(file_path).export(file_path[0:-3] + "mp3", format="mp3")
  os.remove(file_path)
```
